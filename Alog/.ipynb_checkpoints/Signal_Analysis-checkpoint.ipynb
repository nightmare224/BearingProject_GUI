```python
data = dict()
for i, name in enumerate(file_lst):
    # 取得資料
    data[name_lst[i]] = np.load(name)
for name in name_lst:
#     plt.figure(figsize = (20, 5))
#     plt.title(label = '{} in time Domain'.format(name), fontdict = {'fontsize': 15})
#     plt.plot(data[name][30][256:], label = 'time')
#     plt.legend(loc=1,prop={'size': 15})
#     plt.show()
    pass
```

```python
time = data['ball'][10][256:512]
time = (time - np.mean(time)) / np.std(time)
time_smooth = savgol_filter(time, window_length = 5, polyorder = 3)
freq = np.abs(np.fft.fft(time, norm = 'ortho'))
time_envelope = np.abs(hilbert(time))
freq_envelope = np.abs(np.fft.fft(time_envelope, norm = 'ortho'))
```

```python
figsize = (20, 5)

# Plot time envelope
title = 'Time Domain'.format(name)
plt.figure(figsize = figsize)
plt.title(label = title, fontdict = {'fontsize': 15})
plt.plot(time, label = 'time data')
plt.plot(time_smooth, label = 'smooth')
plt.legend(loc=1, prop={'size': 15})
plt.show()


# Plot time envelope
# title = 'Freq Domain'.format(name)
# plt.figure(figsize = figsize)
# plt.title(label = title, fontdict = {'fontsize': 15})
# plt.bar(np.arange(freq[20:].shape[0]), freq[20:], label = 'freq data', width=0.2)
# plt.bar(np.arange(freq[20:].shape[0]), freq_envelope[20:], label = 'envelope', width=0.2)
# plt.legend(loc=1, prop={'size': 15})
# plt.show()
```

```python
freqs, psd = welch(time, fs = 2048)
plt.figure(figsize=(20, 5))
plt.bar(freqs, psd, width = 2)
plt.title('PSD: power spectral density')
plt.xlabel('Frequency')
plt.ylabel('Power')
plt.show()
```

```python
peaks, _ = find_peaks(psd, height = 1)
plt.figure(figsize=(20, 5))
plt.plot(freqs[peaks], psd[peaks], "x")
plt.bar(freqs, psd, width = 0.002, color = 'r')
plt.title('PSD: power spectral density')
plt.xlabel('Frequency')
plt.ylabel('Power')
plt.show()
```
